```python
from qiskit import *
qiskit.__version__
%config InlineBackend.figure_format = 'svg'
```

```python
quantumRegister = QuantumRegister(2)
```

```python
classicalRegister = ClassicalRegister(2)
```

```python
circuit = QuantumCircuit(quantumRegister,classicalRegister)
```

```python
%matplotlib inline
```

```python
circuit.draw()
```

```python
circuit.h(quantumRegister[0])
```

```python
circuit.draw(output="mpl")
```

```python
circuit.cx(quantumRegister[0],quantumRegister[1])
circuit.draw(output="mpl")
```

```python
circuit.measure(quantumRegister,classicalRegister)
```

```python
circuit.draw(output="mpl")
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
res = execute(circuit, backend = simulator).result()
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
print(res.get_counts(circuit))
plot_histogram(res.get_counts(circuit))
```

```python
IBMQ.load_account()
```

```python
provider = IBMQ.get_provider('ibm-q')
```

```python
quantumComputer = provider.get_backend('ibmq_burlington')
```

```python
job = execute(circuit, backend = quantumComputer)
```

```python
from qiskit.tools.monitor import job_monitor
```

```python
job_monitor(job)
```

```python
res = job.result()
```

```python
plot_histogram(res.get_counts(circuit))
```
